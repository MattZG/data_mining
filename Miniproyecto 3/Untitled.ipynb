## 1. Preprocesamiento

Con las librerias listas, ahora importamos la base de datos. En esta ocasión voy a utilizar una base de datos obtenida desde [Kaggle](https://www.kaggle.com/datasets/diraf0/sloan-digital-sky-survey-dr18) que contiene datos observaciones de la Via Lactea, en la cual se clasifican los cuerpos celentes encontrados en 3 tipos:
1. Estrella
2. Galaxia
3. Quasar

Dicho lo anterior, empezamos por importar y vizualizar nuestra base de datos.

```python
df = pd.read_csv('SDSS_DR18.csv')
df
```

Se puede ver que tenemos un total de 43 atributos y 100000 instancias, tal que hay bastante informacion para trabajar. Dado que quiero hacer una clasificación basada unicamente en las caracteristacas propias de cada cuerpo celeste, estas variables no ofrecen información de valor en dicho analisis por lo tanto las voy a eliminar de la base de datos, utilizando el comando *drop*.

Con esto voy a revisar que información restante que contiene la base de datos como el numero de atributos actuales, que tipo de dato poseen y se hay algun datos faltante. Para ello se recurre al metodo *info*.

Con esto podemos sacar dos conclusiones, la primera es que tenemos una base de datos limpia y sin nulos además de que solo posee una variables categorica. Por lo tanto no hay necesidad de realizar un trabajo de limpieza mayor, se puede trabajar con el estado actual del dataframe. 

## 2. Exploración de la base de datos

Ahora empezamos el analisi exploratio de la base datos. Iniciando con ver las caracteristicas estadisticas claves a traves del comando *describe*.

```python
df.describe()
```

Con esta nueva informacion se pueden apreciar algunas caracteristicas de la información. Tal que algunas variables posee una desvacion estandar baja, lo nos puede indicar que estas variables posee los datos mas acumulados que el resto. Tambien algunas variables posee una grande distancia entre sus maximos y minimos, tal que podriamos estar antes la presencia de outliners. Estas conclusiones nos presenta un desafio al momento de definir el metodo de normalización, por lo que primero vamos a vizualizar los datos unicos de cada variable para complementar estas conclusiones. Tenemos un alto numero de valores unicos por variable, tal que es complicado desarrollar una nocion sobre como estan distribuidos los datos o si las conclusiones previas son certeras. Ante esta situacion, otra herramienta que tengo a mano es graficar la informacion de cada variable de graficos de caja y billotes, con ello vizualizamos directamente como se distribuyen los datos además de ayudar a determinar que normalización utilizar. Con la funcion *boxplot* y un ciclo *for* podemos recorrar cada columna de la base de datos para realizar un grafico de ellas.

Mientas que en los graficos donde se forman cajas en en su mayoria muestran una distribucion "homogenea", por asi decirlo, entre los datos. Algo nuevo que aportan estos graficos es que además de los valores atipicos, hay una distancia considerable entre los bigotes y las cajas es decir que los minimos y maximos se encuentras distancias de la mayoria de los datos. Dado todo esto, de cara a normalizar la base de datos el metodo adecuado seria una normalización MinMaxScaler que es lo que se realizara a continuacion.

## 3.Homogenización y Normalización

Antes de normalizar, voy homogeneizar la informacion transformado la variable 'class' a tipo numerico con el metodo *LabelEnconder* y luego aplicar el comando *info* para corroborar que dicha transformación no altero los datos y solo cambio el tipo de la variable. 

```python
label = LabelEncoder()
df['class'] = label.fit_transform(df['class'])
df.info()
```

Ahora se normaliza la base de datos. Primero vamos a sacar la columna de *class* dado que no se normaliza la variable obejtivo, para luego una vez aplicado el metodo se recompora y recuperamos la forma previa que tenia la base de datos pero con el resto de variable normalizadas.

```python
objetivo = df['class']
columnas = df.drop('class', axis = 1)
scaler = MinMaxScaler()
```

```python
df_ = scaler.fit_transform(columnas)
df = pd.DataFrame(df_, columns = columnas.columns)
df['class'] = objetivo
```

```python
df
```

## 4. Elección de matriz de atributos y variable objetivo.

Ya con nuestra base de datos normalizada damos paso a definir que variable queremos estudiar y cuales van a ser los datos que utilizamos para ello. Como mencione en el punto anterior, la variable *class* sera nuestra variable obejtivo, ya que clasifica los cuerpos celestes dentro de 3 tipos: estrella, galaxia y quasar. En el caso de la matriz de atributos, dado que el resto de variable son caracteristicas de los cuerpos observadas por los telescopios estas forman nuestra matriz de atributos. Por medio del comando *train_test_split* dividos la informacion en un 70% de ella para entrenar los modelos y el 30% restante para hacer pruerba de precisión.

```python
x = df.iloc[:, 0:32]
y = df['class']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

## 5. Experimentos de entrenamientos de algoritmos

En este punto vamos a trabajar con los algoritmos de clasificación visto en clases y la precision que tienen. Comento que he decidido realizar el punto 3 y 4 de forma conjunta dado que me parece mas adecuado de cara a discutir resultados.

### 5.1. Algoritmo KNN

Empezamos por trabajar con el algoritmo KNN, el cual busca los k-vecinos, del conjunto de entrenamiento, más cercanos al punto objetivo que se busca predecir. Entre los parametros que voy a manipular se encuentre el numero de vecinos *n_neighbors* donde se indica el numero de vecinos a considerar, *algorithm* que indica el metodo para calcular los vecinos cercanos y *metric* que defini como se determina la distancia de proximidad entre los puntos de datos. Dicho esto, empezamos con el primer experimento, donde utilizamos la version default del algoritmo en la cual el numero de vecinos es 5, el modelo elige de manera automatica cual es mejor metodo para calcular los vecinos cernanos y la metrica es la distancia de *minkowski* como una version generalizada de la distancia euclideana.

```python
model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

Una vez entrenado el modelo, queremos comprobar que tan precisos son los resultados. Primero utilizamos el comando *predict* para generar prediccion a partir de los datos de prueba de la matriz de atributos. Luego con el *accuracy_score* se calcula la precision de las prediccion y el modelo que hemos entrando.

```python
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f'La precisión es de : {acc * 100:.2f}%')
```

Con una precisión de aproximadamente el 93% podemos concluir que la version defalut del algoritmo funciona bien con nuestra base de datos. Para tener una revision mas detallada vamos a recurir a las metricas de evalucion que entrega el comando *classification_report*.

```python
print(classification_report(y_test, y_pred))
```

Vamos mirando punto por punto. En el caso del parametro *precision* vemos que para la clase 1 muestra el menor valor, tal que hay un porcentaje de las predicciones echas por el modelo que son falsos positivos. Este resultado termina teniendo efecto en las siguiente metricas de *recall*, aunque en este caso al considerar falsos negativos el porcentaje es mayor y viendo el *f1-score* que considera ambos valores se obtiene el valor mas bajo de los tres. En conclusion, aunque nuestro primer modelo muestra un algo porcentaje de precision, flaquea cuando se trata de la clase 1.

Ahora probaremos el modelo KNN, pero con nuevos parametros. Vamos a disminuir el numero de vecinos a considerar a 4, ahora voy a elegir el algoritmo de *ball_tree* para calcular los vecinos, además es mas efectiva con base de datos de alta dimensionalidad, y calculando la proximidad con *manhattan* que considera la distancia al sumar la diferencia en valor absoluto entre los puntos. 

Posteriormente calculamos la precision general del modelo con *accuracy_score* y revisaremos en detalle con *classification_report*. 

```python
model = KNeighborsClassifier(n_neighbors= 4, algorithm = 'ball_tree', metric = 'manhattan')
model.fit(X_train, y_train)
```

```python
y_pred_ = model.predict(X_test)
acc = accuracy_score(y_test, y_pred_)
print(f'La precisión es de : {acc * 100:.2f}%')
```

```python
print(classification_report(y_test, y_pred_))
```

Respecto a la precision en general vemos un aumento, lo cual es consistente con que el algoritmo *ball_tree* funciona mejor con altas dimensionalidades, sin embargo en el caso anterior deje que el modelo eligiera el mejor algoritmo para calcular los vecinos, asi que probablemente haya utilizando el mismo algoritmo dada la alta dimensionalidad. Por lo tanto, se puede intuir que este aumento se deba al cambio en el numero de vecinos o la forma de calcular la proximidad. Si ahora analizamos las metricas, hay 3 cambios: en *precision* tanto la clase 0 como la 2 han tenido una disminucion del 1% mientras que la clase 1 aumento en un 3%. Sin embargo, al ver los valores de *recall* y *f1-score* la clase 0 presenta otro aumento del 1% mientras que la clase 2 muestra una disminución en ambas y la clase 2 no presenta cambios. Ante estos resultados, intuyo que el algoritmo tiene mas facilidad para predecir la clase 2, clase 0 y la clase 1 en ese orden. Diria que una posible razon para esto es el numero de informacion que posee para ajustar el modelo, dado que en global la clase 1 posee menos muestras, el algoritmo debe tener facilidad para predecir ciertas clases sobre otras. 

Para comprobar el efecto del algoritmo vamos a cambiar el parametro a *kd_tree* que funciona mejor con datos de baja dimensionalidad, además disminuire el numero de vecinos a considerar y calcular la proximidad con la distancia *eulidean*. Posteriormente calcularemos la precision del algoritmo y veremos el detalle de como predice cada clase.

```python
model = KNeighborsClassifier(n_neighbors= 3, algorithm = 'kd_tree', metric = 'euclidean')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f'La precisión es de : {acc * 100:.2f}%')
```

```python
print(classification_report(y_test, y_pred))
```

Dada que la precision decae a un 93%  algo esperable dado que el algoritmo no funciona bien con altas dimensionalidad, tal que si se enfoca en el numero de vecinos dado que tenemosuna precision similar al primer caso se puede concluir que dentro de los 3 casos no he elegido el numero adecuado de k para dar un aumento significativo al poder predictivo del algoritmo, sin embargo tampoco es que se encuentre muy alejado que en todos los casos tenemos un 90% de precision lo cual no es un resultado menor. Lo mismo se puede decir de la distancia, aunque dado que los valores en este caso junto al primero son en cierto sentido similares me atreveria a decir que la distancia del segundo caso sean el parametro adecuado para el modelo.

Respecto a las metricas cuando observamos los resultados de la clase 0 tenemos buenos resultados, notar que son similares al primer caso, tal que se confirma que el algoritmo en general tiene facilidad para predecir esta clase. Tambien vemos una disminución notable en la clase 1 en comparación a los casos anteriores tal que se confirma que en general el algoritmo KNN tiene dificultades para predecir esta clase. Respecto a la clase 2, hay una disminución menor que se podria considerar esperable dado que este caso del modelo tiene una reducción de la precision en general.

Veamos ahora como se comporta el algoritmo de *DecisionTreeClassifier* en comparación a *KNN* enfretando el mismo problema.

### 5.2. Algoritmo DecisionTree

Este algoritmo contruye un arbol de decisiones basado en las caracteristicas entregadas por los datos para lograr separar las clases que posee nuestra variable objetivo. En este caso voy a modificar los parametros *criterion* que mide la calidad de la separación entre variables, *max_depth* para establecer el nivel de profundidad del arbol al punto de que mientras mas profundo mas datos se utilizan para hacer la clasificación y *min_sample_leaf* donde se indidca el minimo de muestras que se requieren para formar las hojas del arbol.

En el primer caso vamos  fijar los parametros como *gini* que minimizar la impuras de los nodos del arbol, con una profundidad de nivel 2 y con un minimo de muestras de 2. Luego utilizando las mismas metricas que en el caso de *KNN* para medir la precision del modelo en diferentes casos.

```python
model = DecisionTreeClassifier(criterion = 'gini', max_depth = 2, min_samples_leaf = 2)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f'La precisión es de : {acc * 100:.2f}%')
```

```python
print(classification_report(y_test, y_pred))
```

A nivel general este modelo presenta una predicción del 93% lo cual es un resultado buen para empezar, sin embargo al mirar las metricas nuevamente se aprecia que al algoritmo presenta dificultades para predecir la clase 1, similar al caso con *KNN*, tal que como primera conclusión en este caso podriamos decir que al ser la clase con menos datos el algoritmo no tiene suficiente información para predecirla.

Para continuar con la experimentación vamos a modificar los parametros, aumentando la profunidad y el minimo de muestras a 3 además de cambiar el criterio al *entropy* para maximizar la ganancia de información. Luego analizamos los resultados obtenidos al ver las metricas.

```python
model = DecisionTreeClassifier(criterion = 'entropy', max_depth = 3, min_samples_leaf = 3)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f'La precisión es de : {acc * 100:.2f}%')
```

```python
print(classification_report(y_test, y_pred))
```

A nivel general tenemos un ligero aumento en la precision del modelo, particularmente pienso que se debe a que hemos aumentado la profundidad y el numero de muestras, dado mas espacio al modelo para ajustarse a la base de datos y mejorar la precision. Si este fuera el caso en el proximo experimento deberiamos ver otro ligero aumento en la precision. Respecto a las metricas, el analisis es un poco mas complejo ya que si observamos las clase 1 y 2 ambas presentan flucutuaciones en las metricas, mientras una sube otra baja. Aunque considerando que *recall* y *f1-score* se contruyen en base a la metrica anterior nos impusa a observar *precision* vemos un aumento en la clase 1, lo cual es nuevo dentro de este analisis en comparación a los casos anteriores. Aun *recall* presenta una disminución se compensa con el aumento en *f1-score*. Voy a tomar esto como un resultado positivo de cara que en global vemos un aumento en la precison. Respecto a los resultados de la clase 2, el aumento es minimo tal que queda comparar con el ultimo experimento para sacar una conclusion mas relavante.

Finalmente para concluir este experimento vamos nuevamente a aumentar la profundidad y minimo de muestras en 1. Tambien, cambiar al citerio a *log_loss* para medir la discrepancia entre las predicciones y los valores reales. Cuando tengamos los resultados veremos si la conclusion previa va en la dirección correcta.

```python
model = DecisionTreeClassifier(criterion = 'log_loss', max_depth = 4, min_samples_leaf = 4)
model_ = model.fit(X_train, y_train)
```

```python
y_pred = model_.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f'La precisión es de : {acc * 100:.2f}%')
```

```python
print(classification_report(y_test, y_pred))
```

Con este resultado podemos confirmar la conclusion anterior, hemos tenido otro aumento en la precision tal que para este caso mientras mas información le proporcionemos al algoritmo mejores seran las predicciones. Ya no solo a nivel general, las metricas tambien muestran mejoria salvo para la clase 1.

En conclusión, con los resultados obtenidos con ambos algoritmos podemos decir que la clase 1 es la que menos datos posee tal que dificulta a los modelos predecirla, al ser una base de datos de alta dimensionalidad es mejor utilizar el metodo *kd_tree* cuando utilizamos *KNN*, en el caso de utilizar *DecisionTree* mientras más muestras y profundidad tenga mejor sera su poder predictivo.
